# 2.5 统计本月店铺的平均消耗金额/消耗最多的前五家店铺以及所在的省市区

```python
average = df[['店铺名称', '当日消耗', '省', '市', '区', '渠道名称']]
average_g = average.groupby(['店铺名称'], as_index=False)
df_sum = average_g.sum().reset_index()
df_mean = average_g.mean().reset_index()
df_sum['sum'] = df_sum['当日消耗']
df_mean['mean'] = df_mean['当日消耗']

average1 = average.drop_duplicates(['店铺名称'])
average2 = pd.merge(average1, df_sum, how='left', left_on='店铺名称', right_on='店铺名称')
average3 = pd.merge(average2, df_mean, how='left', left_on='店铺名称', right_on='店铺名称')
average4 = average3[['店铺名称', 'sum', 'mean', '省', '市', '区', '渠道名称']].sort_values(by=['sum'], ascending=0)
average4.head()
```
